```python
import pandas as pd
data = pd.read_csv (r"C:\src\db\internship.csv", index_col=False, delimiter = ',') 

```

```python
import mysql.connector as msql
from mysql.connector import Error
```

```python
try:
    conn = msql.connect(host='localhost', user='root', password='root')

    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute('DROP DATABASE IF EXISTS internship;')
        cursor.execute("CREATE DATABASE internship")
        print("Database is created")

except Error as e:
    print("Error while connecting to MySQL", e)
```

```python

try:
    conn = msql.connect(host='localhost', database='internship', user='root', password='root')

    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS internship_data;')
        print('Creating table....') 

        cursor.execute("CREATE TABLE internship_data(company_name varchar(255),company_location varchar(255),region varchar(255),is_shortlisted boolean)")
        print("Table is created....")
        #loop through the data frame
        for i,row in data.iterrows():
            #here %S means string values 
            sql = "INSERT INTO internship.internship_data VALUES (%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```
